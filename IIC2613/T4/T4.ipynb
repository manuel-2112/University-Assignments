#### ¿Qué es el balance/desbalance de clases?

El balance de clases se refiere a una distribución equitativa de instancias entre las diferentes clases de un conjunto de datos. En un conjunto de datos balanceado, cada clase tiene aproximadamente el mismo número de instancias. Por el contrario, en un conjunto de datos desbalanceado, una o más clases tienen significativamente más instancias que otras.

#### ¿Cómo puede afectar el desbalance de clases a un algoritmo clasificador?

El desbalance de clases puede afectar negativamente el rendimiento de un algoritmo clasificador de varias maneras:

1. **Sesgo del Modelo**: Los algoritmos tienden a favorecer la clase mayoritaria, ya que aprenden a minimizar el error clasificando correctamente las instancias de la clase dominante.
2. **Métricas de Evaluación Engañosas**: Las métricas de evaluación como la precisión pueden ser engañosas en conjuntos de datos desbalanceados, ya que un modelo puede tener alta precisión clasificando todas las instancias como la clase mayoritaria.
3. **Desempeño en Clases Minoritarias**: El rendimiento en las clases minoritarias puede ser pobre, resultando en baja sensibilidad y baja precisión para esas clases.

#### Técnicas para Contrarrestar el Desbalance de Clases

1. **Submuestreo de la Clase Mayoritaria**: Reducir el número de instancias de la clase mayoritaria.
2. **Sobremuestreo de la Clase Minoritaria**: Aumentar el número de instancias de la clase minoritaria mediante técnicas como duplicación de datos o generación de nuevas instancias sintéticas.
3. **Ajuste de Pesos**: Ajustar los pesos de las clases en la función de pérdida del modelo para penalizar más los errores en la clase minoritaria.

ref: https://machinelearningparatodos.com/tratamiento-de-clases-desbalanceadas/

En esta sección, deberás llevar a cabo un tratamiento de los datos para garantizar que los clasificadores puedan ser entrenados de manera adecuada. Es importante resaltar que todas las decisiones que tomes deberán estar debidamente justificadas. Para comenzar responde a las preguntas:

¿Por qué es importante preprocesar los datos y cómo puede esto afectar a nuestro problema de clasificación?

El preprocesamiento de datos es una etapa importante en cualquier proyecto de aprendizaje automático ya que consiste en preparar y limpiar los datos para que sean adecuados para el entrenamiento de modelos. La importancia del preprocesamiento se basa en varios factores:

1. **Calidad de los Datos**: Los datos del mundo real generalmente están sucios, incompletos o inconsistentes. El preprocesamiento ayuda a limpiar los datos eliminando o corrigiendo valores atípicos, datos faltantes y errores.

2. **Normalización y Estandarización**: Las características de los datos pueden tener diferentes escalas y unidades. La normalización y estandarización aseguran que todas las características contribuyan de manera equitativa al modelo, mejorando así su rendimiento y convergencia.

3. **Reducción de Dimensionalidad**: El preprocesamiento incluye técnicas como la selección de características y la reducción de dimensionalidad, lo que ayuda a eliminar características irrelevantes o redundantes, reduciendo el tiempo de entrenamiento y mejorando la interpretabilidad del modelo.

4. **Codificación de Variables Categóricas**: Las características categóricas deben ser convertidas a un formato numérico adecuado para que los algoritmos de aprendizaje automático puedan procesarlas.

5. **Manejo de Desbalance de Clases**: En base a lo anterior, el desbalance de clases puede afectar negativamente el rendimiento del modelo. Las técnicas de sobremuestreo, submuestreo o ajuste de pesos pueden ser aplicadas durante el preprocesamiento para mitigar este problema.

En el contexto de este problema de clasificación, el preprocesamiento es esencial para:

- **Mejorar la precisión del modelo**: Datos limpios y bien preparados permiten que los clasificadores aprendan patrones más precisos y generalizables.
- **Evitar sobreajuste**: Eliminando características irrelevantes y normalizando los datos, podemos reducir la complejidad del modelo y prevenir el sobreajuste.
- **Asegurar que el modelo maneje adecuadamente el desbalance de clases**: Aplicando técnicas de balanceo de clases, podemos garantizar que el modelo tenga un rendimiento equilibrado en ambas clases, lo cual es importante para predecir correctamente la posibilidad de tsunami.

### 2.5.2 F1-Score
Investiga qué es la métrica F1-score y calcúlala a partir de los resultados, no puedes usar implementaciones ya existentes para calcularlo. Luego argumenta, ¿qué mide el score F1 en problemas de clasificacion binaria?

```python
# Función para calcular F1-score
def calculate_f1_score(precision, recall):
    return 2 * (precision * recall) / (precision + recall)
```

```python
# Calcular precisión, recall y F1-score para KNN
f1_score_knn = calculate_f1_score(precision_knn, recall_knn)

# Calcular precisión, recall y F1-score para Árbol de Decisión
f1_score_dt = calculate_f1_score(precision_dt, recall_dt)

# Calcular precisión, recall y F1-score para Random Forest
f1_score_rf = calculate_f1_score(precision_rf, recall_rf)

# Calcular precisión, recall y F1-score para SVM
f1_score_svm = calculate_f1_score(precision_svm, recall_svm)


metrics_df['F1-score'] = [f1_score_knn, f1_score_dt, f1_score_rf, f1_score_svm]
# Mostrar la tabla comparativa
metrics_df
```

El score F1 mide el balance entre precisión y recall en problemas de clasificación binaria. Es especialmente útil en conjuntos de datos desbalanceados, ya que proporciona una evaluación equilibrada de la capacidad del modelo para clasificar correctamente las instancias positivas. Un alto score F1 indica un buen desempeño en términos de precisión y recall.

### 2.5.5 Ajuste de hiperparametros
Para el algoritmo con los mejores resultados elige 2 hiperparámetros y entrena 3 modelos variando estos, con este el mismo algoritmo, realiza predicciones con este y luego evalúalos. ¿Qué combinación de hiperparámetros es la con mejores resultados?

```python
# Definir combinaciones de hiperparámetros
hyperparameters = [
    {'n_estimators': 100, 'max_depth': 10},
    {'n_estimators': 200, 'max_depth': 20},
    {'n_estimators': 150, 'max_depth': None},
    {'n_estimators': 200, 'max_depth': None},
]

# Entrenar y evaluar los modelos
results = []

for params in hyperparameters:
    # Crear y entrenar el modelo
    rf = RandomForestClassifier(n_estimators=params['n_estimators'], max_depth=params['max_depth'], random_state=42)
    rf.fit(X_train_scaled, y_train_tsunami)
    
    # Realizar predicciones
    y_pred = rf.predict(X_test_scaled)
    
    # Evaluar el modelo
    accuracy = accuracy_score(y_test_tsunami, y_pred)
    precision = precision_score(y_test_tsunami, y_pred)
    recall = recall_score(y_test_tsunami, y_pred)
    f1 = f1_score(y_test_tsunami, y_pred)
    
    # Guardar los resultados
    results.append({
        'n_estimators': params['n_estimators'],
        'max_depth': params['max_depth'],
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    })

# Crear un DataFrame con los resultados
results_df = pd.DataFrame(results)

# Mostrar los resultados
results_df
```

Es una métrica utilizada para evaluar la precisión de un modelo de regresión. Un menor RMSE indica que las predicciones del modelo están más cerca de los valores reales, reflejando una mejor precisión del modelo. Por el contrario, un mayor RMSE sugiere que las predicciones están más alejadas de los valores reales, indicando un rendimiento menos preciso.

En modelos de clasificación, métricas como la precisión, el recall y el F1-score cumplen una función similar al RMSE en regresión, evaluando la exactitud de las predicciones. Un modelo de regresión con un bajo RMSE puede traducirse en mejores predicciones continuas, lo que puede mejorar los umbrales de clasificación y, por ende, los resultados generales de clasificación.
